```python
df.drop(['Status' , 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()   # to identify null values in file
```

```python
df.dropna(inplace=True)  # for deleting null rows 
```

```python
df.shape   # to check whethter null rows is deleted or not
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.describe()   #provides statistical information regarding our datas
```

```python
#exploratory data operations 
#1. Gender
ax = sns.countplot(x= 'Gender' , data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data = sales_gen )
```

```python
# data operations on age for gender sales
ax = sns.countplot( data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data = sales_age )
```

```python
# data operations on state
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sales_state
sns.barplot(data = sales_state, x='State', y='Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Amount', data= sales_state)
```

```python
ax=sns.countplot(x='Marital_Status', data= df)
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender', data = sales_state)
```

```python
#occupations wise sales purchasing power
df.columns
ax=sns.countplot(x='Occupation', data=df)
sns.set(rc={'figure.figsize':(22,5)})
```

```python
sales_occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_occupation
sns.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

```python
#sales according to product
ax= sns.countplot(x='Product_Category', data=df)
sns.set(rc={'figure.figsize':(25,5)})
```

```python
sales_product=df.groupby(['Product_Category','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sales_product
sns.barplot(x='Product_Category',y='Amount',hue='Gender',data=sales_product)
```
